```python
import numpy as np
import matplotlib.pyplot as plt

def R(t, s, alpha=0.03, APY= 0.03, PR= 0.05):
    return 0.8 - np.exp(-APY * np.sqrt(t) * np.sqrt(s*PR*APY))

# Beispielhafte Werte
t_values = np.linspace(100, 10000, 100)
s_values = np.linspace(0.1, 1000, 100)

# Meshgrid fr 3D-Darstellung
T, S = np.meshgrid(s_values, t_values)
R_values = R(T, S)

# 3D-Darstellung der Funktion
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(T, S, R_values, cmap='viridis')
ax.set_ylabel('BLCS Tokens (t)')
ax.set_xlabel('ETH with APY and PR (s)')
ax.set_zlabel('R(t, s)')
ax.set_title('Rckerstattungsrate R(t, s)')
plt.show()
```
